```python
# EJERCICIO 1.
# Programa que cree un directorio y lea todos los ficheros .csv del directorio actual, y los reescribar en el 
# directorio creado, excluyendo la 1ª linea o cabecera..

import os, csv

os.makedirs('SinCabeceras', exist_ok=True) #creacion del directorio..

for csvFilename in os.listdir('.'): #Listamos los ficheros .csv del directorio actual..
    if csvFilename.endswith('.csv'): # El fichero que sea .csv, lo tratamos..
        print('Ficheros a tratar: ',csvFilename)
        
        csvFilas = [] #Objeto para almacenar cada fila del fichero a tratar en una lista..
        csvFichObj = open(csvFilename) #Objeto/cursor para leer el fichero a tratar..
        readObj = csv.reader(csvFichObj) #Objeto/cursor para leer fichero linea a linea
        
        for indLinea in readObj: # recorremos las lineas del fichero a tratar
            if readObj.line_num == 1: # Saltamos la 1ª linea / cabecera del fichero a tratar..
                continue
            
            csvFilas.append(indLinea) # incorporamos la fila en la lista..
        
        csvFichObj.close() #cerramos el fichero tratado..
        
        # Abrimos un objeto que cree un fichero con el nombre contenido en 'csvFilename', en la carpeta 'SinCabeceras'
        # del directorio actual (si existe y si no lo crea), y sin escribir lineas en blanco..
        csvFichObj = open(os.path.join('SinCabeceras', csvFilename), 'w', newline='')
        writeObj = csv.writer(csvFichObj) # Objeto/cursor para escribir en el fichero a tratar..
        
        for indLinea in csvFilas: # Recorremos la lista obtenida anteriormente, para insertar las lineas en el nuevo fichero
            writeObj.writerow(indLinea)
        
        csvFichObj.close()
```
