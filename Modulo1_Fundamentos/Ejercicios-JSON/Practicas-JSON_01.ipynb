```python
# MÓDULO PARA ARCHIVOS JSON.
# - Está basado en JavaScript
# - Es independiente del lenguaje
# - archivos .json
# - Representas objetos de manera textual mediante parejas clave = valor (diccionarios) encerradas entre {}
# - Para formatear un texto JSON a diccionario Python -> json.loads(DATAjson)
# - Para escribir un diccionario de datos python en JSON -> json.dumps(DATApython)

# - Ejemplo:
#   {
#
#   "Nombre":"Juan Serrano Sanchez",
#
#   "DNI":"4569883R",
#
#   "Edad":45, "Asignaturas matriculadas":
#
#    {
#
#      "Obligatorias": ["Algebra", "Matematicas II", "Geometria"],
#
#      "Optativas":["Seminario I", "Seminario II"],
#
#      "Libre Eleccion": ["Informatica", "Musica"]
#
#    }
#
#   }

# Programa que pide una petición url a la API de starWars recuperando un texto en formato JSON y lo trata para mostrar
# la información relevante de los personajes..

# Ejemplo de formato JSON de un personaje SW a través de la API..
# {
#	"name": "Luke Skywalker",
#	"height": "172",
#	"mass": "77",
#	"hair_color": "blond",
#	"skin_color": "fair",
#	"eye_color": "blue",
#	"birth_year": "19BBY",
#	"gender": "male",
#	"homeworld": "http://swapi.dev/api/planets/1/",
#	"films": [
#		"http://swapi.dev/api/films/1/",
#		"http://swapi.dev/api/films/2/",
#		"http://swapi.dev/api/films/3/",
#		"http://swapi.dev/api/films/6/"
#	],
#	"species": [],
#	"vehicles": [
#		"http://swapi.dev/api/vehicles/14/",
#		"http://swapi.dev/api/vehicles/30/"
#	],
#	"starships": [
#		"http://swapi.dev/api/starships/12/",
#		"http://swapi.dev/api/starships/22/"
#	],
#	"created": "2014-12-09T13:50:51.644000Z",
#	"edited": "2014-12-20T21:17:56.891000Z",
#	"url": "http://swapi.dev/api/people/1/"
# }

import json, urllib

apiServiceUrl = 'https://swapi.dev/api/people/'

while True:
    opc = input('Introduce un valor de personaje del [1..82] o [0] para salir: ')
    
    if int(opc) <= 0:
        break
    else:
        urlApi = apiServiceUrl + opc + '/?format=json'
        print('Recuperando [',urlApi,']')
        
        request1 = urllib.request.Request(urlApi)
        request1.add_header('User-Agent','cheese') # PORQUE "USER-AGENT" Y "CHEESE"??...
        
        datos = urllib.request.urlopen(request1).read() # Instrucción para leer y pasar los datos a una variable
        print(datos)
        print('Recuperados ',len(datos),' caracteres.')
        
        # Realizamos un tratamiento de errores del módulo urllib y json..
        try:
            pythonData = json.loads(datos) #convertimos o transformamos los datos en variable python (diccionario)
        except:
            pythonData = None
        print(pythonData)
        
        # Seleccionamos los valores del diccionario, en función de su clave..
        print ("Nombre: ",pythonData['name'])
        print ("Altura: ",pythonData['height'])
        print ("Color de cabello: ",pythonData['hair_color'])
        print ("Color de piel: ",pythonData['skin_color'])
        print ("Color de ojos: ",pythonData['eye_color'])
        print ("Año de nacimiento: ",pythonData['birth_year'])
        print ("Genero: ",pythonData['gender'])

        
        
    
        
```

```python
# Programa para recuperar los datos asociados a una IP pública, mediante el servicio o API en formato JSON
# de GEOIP "http://ip-api.com/"


import urllib, json

apiServiceUrl = "http://ip-api.com/json/"

while True:
        
    opc = input('Introduce la IP a consultar o [Fin] para salir: ')
        
    if opc == 'Fin':
        break
    else:
        urlApi = apiServiceUrl + opc
        print('Recuperando [',urlApi,']')
        
        request1 = urllib.request.Request(urlApi)
        request1.add_header('User-Agent',"Cheese")
        
        datos = urllib.request.urlopen(request1).read()
        print(datos)
        print('Recuperados ',len(datos),' caracteres.')
        
        try:
            pythonData = json.loads(datos)
        except:
            pythonData = None
        
        if ( "status" in pythonData and pythonData['status'] == 'fail' ): #validamos el valor de la clave "status" de la petición
            print('Error al obtener datos: ',pythonData['message']) # en el valor de la clave "message", está el error..
        else:
            print(pythonData)
            
            print ("Ciudad: ",pythonData['city'])
            print ("Población: ",pythonData['country'])
            print ("CP: ",pythonData['countryCode'])
            print ("isp: ",pythonData['isp'])
            print ("latitud: ",pythonData['lat'])
            print ("longitud: ",pythonData['lon'])
            print ("Región: ",pythonData['region'])
            print ("Nombre Región: ",pythonData['regionName'])
            print ("Zona Horaria: ",pythonData['timezone'])
```
