```python
# MÓDULO PARA FICHEROS XML.

#Para ilustrar el uso de XML, supongamos que se quiere representar mediante un documento XML la siguiente información:
#Persona 1:

#Nombre: Roberto Casas

#Email: ro.casas@direccion.com

#Amigos: Leire, Pepe

#Persona 2:

#Nombre: Leire García

#Email: le.gracia@direccion.com,le.garcia@hotmail.com

#Amigos: Ricky

#Persona 3:

#Nombre: José Manzaneda

#Email: j.manzaneda@direccion.com , jman@hotmail.com

#Amigos: Ricky

#Enemigos: Leire

#<?xml version="1.0" encoding="UTF-8" ?>
#<listin>
#	<persona sexo="hombre" id="ricky">
#		<nombre>Roberto Casas</nombre>
#		<email>ro.casas@direccion.com</email>
#		<relacion amigo de="leire pepe"/>
#	</persona>
#	<persona sexo="mujer" id="leire">
#		<nombre>Leire Garcia</nombre>
#		<email>le.garcia@direccion.com</email>
#		<email>le.garcia@hotmail.com</email>
#		<relacion amigo_de="ricky"/>
#	</persona>
#	<persona sexo="hombre" id="pepe">
#		<nombre>Jose Manzaneda</nombre>
#		<email>j.manzaneda@direccion.com</email>
#		<email>jman@hotmail.com</email>
#		<relacion enemigo_de="leire" amigo_de="ricky"/>
#	</persona>
#</listin>


# Procesamiento SAX (Simple API for XML):

# - Interfaz orientada a eventos, que permite leer el contenido como una secuanecia de datos e interpretar
#   las etiquetas según se van encontrando..
# - Según se va analizando el documento se crean los comentados 'Eventos de análisis', xejmeplo al principio
#   de una etiqueta o elemento, se crea un evento, y al final de esta otro evento
# - Para controlar estos eventos, existen funciones denominadas 'Manejadoras de eventos', que invocandolas cuando
#   se produce un evento, nos proporciona la posibilidad de realizar la acción sobre ese elemento.
# - La info ya procesada no se puede manipular, para poder hacerlo o se guarda en una estructura de datos o se
#   vuelve vuelve a procesar..

# - Ejemplo: Para este archivo se crearian los siguentes eventos..
# <?xml version=“1.0”?> Inicio de documento
#    <doc> Inicio elemento doc
#         <par> Inicio elemento par
#         Hola Mundo # Caracteres..
#         </par> Fin elemento par
#     </doc> Fin elemento doc
# Fin documento

# - Manejador CONTENTHANDLER y sus Métodos:

# <?xml version="1.0" encoding="UTF-8" ?>
# <Datos>
# <Libro isbn="0-678-12345-9">
#   <titulo>Libro 1</titulo>
#   <fecha>Diciembre 2001</fecha>
#   <autor>Pepito Perez</autor>
# </Libro>
# …
# </Datos>

import xml.sax

class manejadorCatalogo (xml.sax.ContentHandler): # Creamos la clase manejadora con sus métodos..
    def __init__(self):
        self.datos=""
        self.titulo=""
        self.fecha=""
        self.autor=""
        
    def startElement (self,etiqueta,attr): # Método inicial para el inicio de cada elemento..para esta caso
                                           # solo queremos tratar las etiquetas "Libro"..
        self.datos = etiqueta
        # if self.datos == 'Libro' # se podría poner asi tb..
        if etiqueta == "Libro":
            print("****Libro****")
            isbn = attr["isbn"]
            print("isbn: ",isbn)
    
    def endElement (self,etiqueta): # Método para el final de los elementos..para este caso tratamos las
                                    # etiquetas ...., para que pinten el contenido de la etiqueta..
        if self.datos == "titulo":
            print("Título: ",self.titulo)
        elif self.datos == "fecha":
            print("Fecha: ",self.fecha)
        elif self.datos == "autor":
            print("Autor: ",self.autor)
        
        self.datos = ""
    
    def characters (self,contenido): # Método para agregar en una variable el contenido de las etiquetas
                                     # en este caso tratamos las etiquetas ...
        if self.datos == "titulo":
            self.titulo = contenido
        elif self.datos == "fecha":
            self.fecha = contenido
        elif self.datos == "autor":
            self.autor = contenido

if __name__ == "__main__": # Creamos la parte principal del programa..
    
    parser1 = xml.sax.make_parser() # Creamos el parser para leer el doc XML..
    parser1.setFeature(xml.sax.handler.feature_namespaces,0) # Inhabilitamos los "namespaces"..
    
    handler1 = manejadorCatalogo() # Creamos nuestro manejador invocando a la clase..
    
    parser1.setContentHandler(handler1) # invocamos al método set..
    parser1.parse("datos.xml") # Parseamos la info del documento XML..
```

```python
# Procesamiento DOM (árbol DOM)(Document Object Model):

from xml.dom.minidom import parse
import xml.dom.minidom

arbolDom = xml.dom.minidom.parse("datos.xml") #método "parse" del objeto "minidom", obtenemos el fichero xml en variable..
Libros = arbolDom.getElementsByTagName("Libro") # Obtenemos una lista de todos los elementos/etiquetas "Libro", y su contenido
print(arbolDom)
print(Libros)
for indDom in Libros: # recorremos la lista..
    print("****Libro****")
    isbn = indDom.getAttribute('isbn')
    titulo = indDom.getElementsByTagName('titulo')[0] # Para este tipo de nodo se utiliza [0], xa recuperar contenido, 
    fecha = indDom.getElementsByTagName('fecha')[0]   # en vez del elemento de memoria del árbol DOM..
    autores = indDom.getElementsByTagName('autor') # Al poder ser más de uno no se pone el [0]..
    print('ISBN:', isbn)
    print('Titulo:', titulo.firstChild.data)
    print('Fecha:', fecha.firstChild.data)
    for indAutor in autores:
        print('Autor:', indAutor.firstChild.data)
    print('*************')

    
arbolDom2 = xml.dom.minidom.parse('rss.xml')
channel = arbolDom2.getElementsByTagName("channel")

for indChannel in channel:
    print("**** Channel *****")
    print("Titulo Canal:", indChannel.getElementsByTagName("title")[0].firstChild.data)
    print("Link Canal:", indChannel.getElementsByTagName("link")[0].firstChild.data)
    print("Descripción canal:", indChannel.getElementsByTagName("description")[0].firstChild.data)
    print("--- Noticias Canal -----")
    items = indChannel.getElementsByTagName("item")
    for indItems in items:
        print("Titulo Noticia:", indItems.getElementsByTagName("title")[0].firstChild.data) 
        print("Link Noticia:", indItems.getElementsByTagName("link")[0].firstChild.data)
        print("Descripción Noticia:", indItems.getElementsByTagName("description")[0].firstChild.data)
        print("---------")
    print("*****************")
```

```python
# Procesamiento ELEMENTTREE.
# - Caracteristicas similares a DOM
# - El árbol generado está formado por objetos de tipo "element", con atributos como "nombre,dicc de atributos,
#   valor textual, secuencia de elementos hijo"

from xml.etree import ElementTree

fich1 = open("datos.xml","rt") # Sintaxis para obtener el árbol del XML..
arbolEtree = ElementTree.parse(fich1)
print(arbolEtree)

for indNodo in arbolEtree.iter(): # Sentencia para mostrar e iterar los elementos del árbol XML..
    print('Nodo tag:',indNodo.tag,'Nodo Text:',indNodo.text,'Nodo Attr:',indNodo.attrib)

ind = 1
for indNodo in arbolEtree.iter("Libro"): # Iterar sobre un elemento en concreto..
    isbn = indNodo.attrib.get("isbn") #sintaxis para recuperar el atributo de un tipo de elemento en concreto..
    print(indNodo.tag, ind,'con ISBN:',isbn)
    ind += 1

raiz = arbolEtree.getroot() # recuperar los primer elemento padre o raiz e iterar en su hijo..
print(raiz)
ind2 = 1
for indNodoHijo in raiz:
    print(indNodoHijo.tag, ind2,'--',indNodoHijo.attrib)
    ind2 += 1

# Acceso por indexación..
print('Título:',raiz[0][0].text,', Fecha:',raiz[0][1].text)
print('Título:',raiz[1][0].text,', Fecha:',raiz[1][1].text)
print('Título:',raiz[2][0].text,', Fecha:',raiz[2][1].text)
print('Título:',raiz[3][0].text,', Fecha:',raiz[3][1].text)
print('Título:',raiz[4][0].text,', Fecha:',raiz[4][1].text)
print('Título:',raiz[5][0].text,', Fecha:',raiz[5][1].text)

# Acesso mediante métodos del objeto ".parse()":
# - find(): recupera el primer subelemento del elemento actual encajando con la descripción dada.
# - findall(): recupera todos los subelementos del elemento actual encajando con la descripción dada.
# - iterfind(): recupera todos los elementos encajando con la descripción dada.
# - text: accede al contenido textual de un elemento.
# - get(atributo): accede al atributo dado del elemento.

ind3 = 1
for indNodo in arbolEtree.findall("./Libro/titulo"):
    print("Titulo", ind3,':',indNodo.text)
    ind3 += 1
```

```python
# Esta API permite realizar un procesamiento basado en eventos como en (SAX), con el método "iterparse()"
# - Genera eventos "start", al inicio del elemento
# - Genera eventos "end", al final del elemento
# - Los datos o el valor textual de cada elemento se puede extraer durante el "parseo"

# Programa que realiza el mismo procesamiento que el de SAX pero con ELEMENTTREE

from xml.etree.ElementTree import iterparse

for (event,element) in iterparse("datos.xml",("start","end")):
    if event == "start":
        if element.tag == "Libro":
            print('****Libro****')
            print('ISBN:', element.attrib["isbn"])
    if event == "end":
        if element.tag == "titulo":
            print('Título:',element.text)
        if element.tag == "fecha":
            print('Fecha:',element.text)
        if element.tag == "autor":
            print('Autor:',element.text)

# Procesar una cadena, que representa un doc XML con el método "fromstring()" de ELEMENTTREE..

cadenaXML = '''<Datos>
                <Libro isbn="0-596-00128-2">
                    <titulo>Python y XML</titulo>
                    <fecha>Diciembre 2001</fecha>
                    <autor>Pepito Perez</autor>
                </Libro>
                <Libro isbn="0-596-15810-6">
                    <titulo>Programacion avanzada de XML</titulo> 
                    <fecha>Octoubre 2010</fecha>
                    <autor>Juan Garcia</autor>
                </Libro>
               </Datos>'''

arbol = ElementTree.fromstring(cadenaXML)
listaNodo = arbol.findall("Libro")

ind4 = 1
for indN in listaNodo:
    print('****Libro',ind4,'****')
    print('ISBN:',indN.get("isbn"))
    print('Título:',indN.find("titulo").text)
    print('Fecha:',indN.find("fecha").text)
    print('Autor:',indN.find("autor").text)
    ind4 += 1
```

```python
# Métodos para la modificación del arbol ELEMENTTREE
# - "element.text" cambiar el contenido o valor de un elemento
# - "element.set" añadir o modificar atributos del elemento o etiqueta
# - "element.append" añadir nuevos hijos
# - *** Con el método "ElementTree.write()" se escribe el nuevo documento

# Programa que modifica el archivo datos.xml, le añade los atributos al elemento "libro"
# "orden" y "ejemplares", y añade el elemento hijo "editorial"..

from xml.etree import ElementTree as et

fich2 = open("datos.xml")
arbol2 = et.parse(fich2)

ind5 = 1
for indNod in arbol2.iter("Libro"): # Nos posicionamos en los elementos "Libro", para modificar su contenido
    ind5str = str(ind5)
    indNod.set("orden",ind5str) # Añadiendo el atributo ("nom Atributo", "valor")
    indNod.set("ejemplares",'Si')
    newElement = et.Element("editorial")
    newElement.text = 'Anaya'
    indNod.append(newElement)
    ind5 += 1

arbol2.write("datos2.xml")
    
    
# Borrar elementos "libro" y sus hijos, que en su attributo "orden" sea igual a "3"..

fich3 = open("datos2.xml")
arbol2 = et.parse(fich3)

raiz = arbol2.getroot() # al analizar edoc XML, vemos que podemos procesar el elemento "libro" como un raiz
for indNod in raiz.iter("Libro"):
    orden = int(indNod.get("orden")) # recuperamos el valor del atributo "orden" y lo convertimos a int..
    if orden == 3:
        raiz.remove(indNod) #eliminamos el Nodo completo con sus hijos..

arbol2.write("datos3.xml")
```
